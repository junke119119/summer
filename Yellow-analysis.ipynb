```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data= pd.read_excel('..\\data\\data_all.xlsx')
```

```python
data
```

```python
## get a reduced data file which contains only values from start to end
pids = data['Patient ID'].unique()
subject=[]
for pid in pids:
    data1 = data[data['Patient ID'] == pid]
    start = data1['Start'].values[0]
    end = data1['End'].values[0]
    data1 = data1.iloc[start-1:end]
    subject.append(data1)
subject = pd.concat(subject)

```

```python
subject_new = subject.groupby('Patient ID')
subject_new
```

```python
subject_new.get_group(1)['Time (min)'].values[0]
```

```python
from datetime import datetime, timedelta
deltatime = []
starttime = []
endtime =[]
for pid in pids:
    start_time = datetime.strptime(subject_new.get_group(pid)['Time (min)'].values[0],"%M:%S")
    delta_start = timedelta(minutes=start_time.minute, seconds=start_time.second)
    end_time = datetime.strptime(subject_new.get_group(pid)['Time (min)'].values[-1],"%M:%S")
    delta_end = timedelta(minutes=end_time.minute, seconds=end_time.second)
    starttime.append((delta_start).seconds)
    endtime.append((delta_end).seconds)
    deltatime.append((delta_end-delta_start).seconds)
```

```python
plt.hist(deltatime)
```

```python
np.mean(deltatime)
```

```python
np.std(deltatime)
```
